# Age Outliers

```python
plt.boxplot(df["Fare"])
plt.title("boxplot of Fare")
```

```python
df["Fare"].describe()
```

```python
q1_F = df["Fare"].quantile(.25)
q3_F= df["Fare"].quantile(.75)
iqr_F = q3_F-q1_F
low_lim_F = q1_F-1.5*iqr_F
low_lim_F
```

```python
up_lim_F = q3_F+1.5*iqr_F
up_lim_F
```

```python
df[df['Fare'] > up_lim_F]
```

```python
df[df['Fare'] < low_lim_F]
```

```python
df.loc[df['Fare']>up_lim_F,'Fare'] = up_lim_F
```

```python
df[df['Fare'] > up_lim_F]
```

```python
plt.boxplot(df["Fare"])
plt.title("boxplot of Fare")
```

here low there are no outliers in low limit, we only had outliers with up limit, and the uplimit was 65.76, so all the people with more than this fare can be considered as one group. so i am changing the values to up limit.
